```python
import pandas as pd
import numpy as np
import scanpy as sc
import os
import json

from matplotlib import pyplot as plt
import seaborn as sns

from ncfs import distances
figdir = os.path.join('figures', 'subselect')
if not os.path.exists(figdir):
    os.makedirs(figdir)
```

```python
subsample = pd.read_csv('data/subsample.csv')\
              .rename(columns={'Unnamed: 0': 'Cells'})
# subsample.index.name = 'cells'
subsample.dropna(axis=1, inplace=True)
```

```python
rc = {"xtick.major.pad": 0,
      "ytick.major.pad": 0,
      "xtick.major.size": 0,
      "xtick.minor.size": 0,
      "ytick.major.size": 0,
      "ytick.minor.size": 0,
      "lines.linewidth": 2,
      "figure.subplot.left": 0.08,
      "figure.subplot.right": 0.95,
      "figure.subplot.bottom": 0.07,
      "figure.subplot.wspace": 0.25,
      "figure.subplot.hspace": 0.25,
      "legend.labelspacing": 0,
      "legend.handletextpad": 0,
      "legend.columnspacing": 0,
      "font.family": "sans-serrif"}
with sns.plotting_context("paper", font_scale=1.25, rc=rc):
    fig, ax = plt.subplots(figsize=(3,3), dpi=200)
    ax = sns.lineplot(data=subsample, x='Cells', y='ARI', style='Dataset',
                      hue='Dataset',
                      lw=3)
    ax.legend_.set_bbox_to_anchor((1, 1))
    for spine in ['top', 'right']:
        ax.spines[spine].set_visible(False)
    plt.savefig(os.path.join(figdir, 'subsample_performance.svg'),
                dpi=200, bbox_inches='tight')
```
